```python
def YesNoto01(data, column_name):
  for i in range(len(data)):
    if data[column_name][i] == "Yes":
      data[column_name][i] = 1
    elif data[column_name][i] == "No":
      data[column_name][i] = 0
```

```python
train_dummy = pd.get_dummies(for_dummy_train)
test_dummy = pd.get_dummies(for_dummy_test)
```

```python
train_new = train.drop(["area_cluster", "segment", "model", "fuel_type", "max_torque", "max_power", "engine_type", "steering_type"], axis = "columns")
test_new = test.drop(["area_cluster", "segment", "model", "fuel_type", "max_torque", "max_power", "engine_type", "steering_type"], axis = "columns")
```

```python
x_train = train_new.drop(["is_claim"], axis="columns")
x_train = pd.concat([x_train, train_dummy], axis=1)
y_train = train_new["is_claim"]
x_test = test_new
x_test = pd.concat([x_test, test_dummy], axis=1)
```
